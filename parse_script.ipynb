```python
import re
import pandas as pd
pattern = r'([a-zA-Z]+):(.+)'
```

```python
data = {
    'name': [],
    'line': []
}
```

```python
with open('witcher3.txt', 'rt') as file:
  for line in file.readlines():
    match = re.findall(pattern, line)
    if match:
      name, line = match[0]
      data['name'].append(name)
      data['line'].append(line)
```

```python
df = pd.DataFrame(data)
```
